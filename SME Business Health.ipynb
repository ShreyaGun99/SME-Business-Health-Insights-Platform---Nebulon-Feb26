```python
import streamlit as st
import pandas as pd

df = pd.read_csv('final_sme_data.csv')
```

```python
st.title("SME Business Health Dashboard")

st.metric("Total Revenue", round(df['Revenue'].sum(),2))
st.metric("Total Profit", round(df['Profit'].sum(),2))
st.metric("Avg Profit Margin", round(df['Profit_Margin'].mean(),2))
```

```python
latest_status = df.iloc[-1]['Health_Status']
st.subheader(f"Business Health: {latest_status}")
```

```python
st.line_chart(df.set_index('date')[['Revenue', 'Total_Expenses']])
st.line_chart(df.set_index('date')['Profit'])
```

```python
if df['Expense_Ratio'].iloc[-1] > 0.8:
    st.warning("⚠️ Expenses are too high compared to revenue")

if df['Anomaly'].iloc[-1]:
    st.error("🚨 Anomaly detected in recent performance")
```

```python
st.write("📌 Business is showing moderate stability with occasional expense spikes.")
```
